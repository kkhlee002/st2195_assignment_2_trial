# Introduction
 In this Jupyter notebook we want to scrap the csv example on cars in the Wikipedia page and save the resultinig output in the local folder

## Load the library

```python
from bs4 import BeautifulSoup
import requests
```

# Get the csv example

Load in the html

```python
webpage_html = requests.get("https://en.wikipedia.org/wiki/Comma-separated_values")
soup = BeautifulSoup(webpage_html.text, 'html.parser')
soup
```

Get the csv example under the header "Example"

```python
Ex_section = soup.find(id='Example') # find the id="Example"
csv_table = Ex_section.findNext('pre').text # Locate it to the "pre" tage
csv_table
```

# Save the csv example into a csv file

```python
f = open('car.csv', 'w')
f.write(csv_table)
f.close()
```
